```python
def change_column(cols):
    Age=cols[0]
    Pclass=cols[1]
    
    
    if pd.isnull(Age):
        if Pclass ==1:
            return 38
        elif Pclass ==2:
            return 29
        else :
            return 25
    else:
        return Age
```

```python

df['Age']=df[['Age','Pclass']].apply(change_column,axis=1)
df1['Age']=df1[['Age','Pclass']].apply(change_column,axis=1)
```
